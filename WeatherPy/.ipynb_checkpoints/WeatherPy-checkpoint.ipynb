# WeatherPy


```python
import requests
import json
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
from citipy import citipy
import numpy as np
from scipy.stats import linregress

from api_keys import g_key
from api_keys import weather_api_key
```

### Generate City List

```python
cities=[]
latitude_range=(-90, 90)
longitude_range=(-180, 180)
latitude_longitude=[]

Latitude = np.random.uniform(low=-90, high=90, size=1500)
Longitude = np.random.uniform(low=-180, high=180, size=1500)
latitude_longitude = zip(Latitude, Longitude)

for lat_long in latitude_longitude:
    city = citipy.nearest_city(lat_long[0], lat_long[1]).city_name
    if city not in cities:
        cities.append(city)


len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it's being processed (with the city number and city name).


```python
    url = "http://api.openweathermap.org/data/2.5/weather?"
    units = "imperial"
    query_url = f"{url}appid={weather_api_key}&units={units}&q="
```

```python
city_name=[]
country=[]
lat=[]
lng=[]
temp=[]
humidity=[]
cloudiness=[]
wind_speed= []
```

```python

count=1

for city in cities:

    try:

        response=requests.get(query_url + city).json()
        city_name.append(response['name'])
        country.append(response['sys']['country'])
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        temp.append(response['main']['temp_max'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        wind_speed.append(response['wind']['speed'])
        print(f"The name of city number {count} is {city}.")
        count=count+1

    except KeyError:
        print(f"City not found, moving on to the next city")
        count=count+1

  
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
weather_df = pd.DataFrame({ "City":city_name,
                            "Country":country,
                            "Latitude":lat,
                            "Longitude":lng,
                            "Temperature (F)":temp,
                            "Humidity (%)":humidity,
                            "Cloudiness (%)":cloudiness,
                            "Wind Speed (mph)":wind_speed})

pd.DataFrame.to_csv(weather_df, 'weather_df.csv')

weather_df.head()    
```

## Linear Regression

```python
northern_hemi_df = weather_df.loc[(weather_df["Latitude"] >= 0)]
southern_hemi_df = weather_df.loc[(weather_df["Latitude"] < 0)]
northern_hemi_df.head()
```

```python
def plot_linear_regression(x_values, y_values, title, text_coordinates):
    
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
    regress_values = x_values * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

    plt.scatter(x_values,y_values)
    plt.plot(x_values,regress_values,"r-")
    plt.annotate(line_eq,text_coordinates,fontsize=15,color="red")
    plt.xlabel('Latitude')
    plt.ylabel(title)
    print(f"The r-value is: {rvalue**2}")
    plt.show()
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = northern_hemi_df["Latitude"]
y_values = northern_hemi_df["Temperature (F)"]
plot_linear_regression(x_values, y_values, 'Temperature (F)', (-55, 90))
```

#### *There appears to be a negative correlation between latitude and max. temp in the Northern Hemisphere.

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = southern_hemi_df["Latitude"]
y_values = southern_hemi_df["Temperature (F)"]
plot_linear_regression(x_values, y_values, 'Temperature (F)', (-55, 90))
```

#### *There appears to be a positive correlation between latitude and max. temp in the Southern Hemisphere.
